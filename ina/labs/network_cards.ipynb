# iMDB actors collaboration network
Graph is available [here](networks/collaboration_imdb.net).

This is a 2004 snapshot of the actor collaboration network derived from the Internet Movie Database (IMDb). In this network, the nodes represent actors, and a connection between two nodes implies that the corresponding actors have shared screen time in at least one movie. Therefore, this representation is a one-mode projection derived from the bipartite graph that initially links actors with their respective film appearances. Note that this graph only contained a few lead actors for each movie, not all of them.

Edge represents **collaboration**.

Node represents **actor**.
